```python
if torch.cuda.is_available():
    device = torch.device("cuda:0")
    print("Cuda Activated")
else:
    device = torch.device("cpu")
    print("Cuda Not Activated")
```

```python
data = torch.tensor(pcr_df.values.astype(np.float32), dtype=torch.float32).to(device)
data = torch.reshape(data, (-1, 1, 40))
target = torch.tensor(target_arr, dtype=torch.float32).to(device)

if torch.cuda.is_available():
    data, target = data.cuda(), target.cuda()
```

```python
#Definition of model
import torch.optim as optim

class Net(nn.Module):
    def __init__(self):
        super(Net, self).__init__()
        
        self.conv1d_layer = nn.Sequential(
            nn.Conv1d(1, 2, 3), #2*38
            nn.ReLU(),
            
            nn.MaxPool1d(2),#19
            
            nn.Conv1d(2, 4, 2), #4*18
            nn.ReLU(),
            
            nn.MaxPool1d(2),#9
            
            nn.Conv1d(4, 8, 2), #8*8
            nn.ReLU(),
            
            nn.MaxPool1d(2),#4
            
            nn.Conv1d(8, 16, 2), #8*3
            nn.ReLU(),
        )
        
        self.fc_layer = nn.Sequential(
            nn.Linear(48, 200),
            nn.ReLU(),
            nn.Linear(200, 40),
            nn.ReLU(),
            nn.Linear(40, 1),
        )
        
    def forward(self, x):
        x = self.conv1d_layer(x)
        
        x = x.view(x.shape[0], -1)
        
        x = self.fc_layer(x)

        return x
```

```python
from torch.optim import Adam
from torch.nn import L1Loss, MSELoss

loss_func = MSELoss()
```

```python
TOTAL_DATA_SIZE = len(data)
BATCH_SIZE = 40
TRAIN_SIZE = int(BATCH_SIZE/5 * 4)
TEST_SIZE = int(BATCH_SIZE/5)
ITER_NUM = 1000
#BATCH_ITER_NUM = 500
SEARCH_ITER_NUM = 20
LEARNING_RATE = 0.0001
result_list=[]; [result_list.append([]) for x in range(len(data))]
print()
```

```python
def batch_iter_search(biter_num_list, train, test):
    train_data, train_target = train
    test_data, test_target = test
    losses = []
    for biter_num in biter_num_list:
        model = Net()
        if torch.cuda.is_available():
            model.cuda()
        optimizer = Adam(model.parameters(), lr = 0.0001)
        
        model.train()
        for batch_iter in range(1, biter_num+1):
            optimizer.zero_grad()
            output = model(train_data)
            loss = loss_func(output, train_target.view(-1, 1))
            loss.backward()
            optimizer.step()
            
        model.eval()
        output = model(test_data)
        loss = loss_func(output, test_target.view(-1, 1))
        losses.append(loss)
        
    biter_num = 0
    min_loss = 100
    for i, loss in enumerate(losses):
        if min_loss > loss:
            min_loss = loss
            biter_num = biter_num_list[i]
    return biter_num

batch_nums = []
for _iter in range(1, ITER_NUM+SEARCH_ITER_NUM+2):
        
    suffled_index = np.random.permutation(TOTAL_DATA_SIZE)[:BATCH_SIZE]
    batch_X  = data[suffled_index]
    batch_y  = target[suffled_index]
    
    train_data, test_data = torch.split(batch_X, [TRAIN_SIZE, TEST_SIZE], dim=0)
    train_target, test_target = torch.split(batch_y, [TRAIN_SIZE, TEST_SIZE], dim=0)
    
    if _iter < SEARCH_ITER_NUM+1: 
        BATCH_ITER_NUM = batch_iter_search([500, 750, 1000, 1500, 2000], (train_data, train_target), (test_data, test_target))
        batch_nums.append(BATCH_ITER_NUM)
        continue
    BATCH_ITER_NUM = int(np.mean(batch_nums))
    
    if _iter == SEARCH_ITER_NUM+1:
        print(BATCH_ITER_NUM)

  #set model parameters & optimizer
    model = Net()
    if torch.cuda.is_available():
        model.cuda()
    optimizer = Adam(model.parameters(), lr = 0.0001)
  
    train_loss = 0.0
    #######################
    #     Train Start     #
    #######################
    model.train()
    for batch_iter in range(1, BATCH_ITER_NUM+1):
        optimizer.zero_grad()
        output = model(train_data)
        loss = loss_func(output, train_target.view(-1, 1))
        loss.backward()
        optimizer.step()
        train_loss = loss
    #######################################################
    
    #######################
    #     Test Start      #
    #######################
    model.eval()
    output = model(test_data)
    loss = loss_func(output, test_target.view(-1, 1))
    output = output.view(-1).tolist()
    print(f'{_iter} train_loss:{train_loss}, test_loss:{loss}')
    
    for i, val in enumerate(suffled_index[-7:]):
        result_list[val].append(output[i])
```

```python
#Function to calculate Ct
def calcCt(lDb, lth):
    ct = np.zeros(len(lDb.columns))
    ic = 0
    
    lDb_over10 = lDb[10:]#baseline은 제외
    
    for c in lDb.columns:
        tmp=lDb_over10.loc[lDb_over10[c]>lth]
        
        if tmp.empty:
            ct[ic]=-1
        else:
            id = tmp[c].index[0]
            rv = lDb[c].loc[id]
            lv = lDb[c].loc[id-1]
            dv = rv-lv
            ct[ic] = id - (rv-lth)/dv
        ic += 1
        
    return ct
```
