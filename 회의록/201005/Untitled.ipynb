```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import display

%matplotlib inline

plt.rcParams["figure.figsize"] = (20,10)
plt.rcParams['axes.grid'] = True 
```

```python
df = pd.read_csv('농도테스트(200904).csv')
```

```python
display(df)
df.columns
```

```python
sns.lineplot(x=df.index, y='x4_10^5', data=df, dashes=False)
sns.lineplot(x=df.index, y='x4_10^5.1', data=df, dashes=False)
sns.lineplot(x=df.index, y='x4_10^5.2', data=df, dashes=False)
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```

```python
sns.lineplot(x=df.index, y='x3_10^5', data=df, dashes=False)
sns.lineplot(x=df.index, y='x3_10^5.1', data=df, dashes=False)
sns.lineplot(x=df.index, y='x3_10^5.2', data=df, dashes=False)
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```

```python
lam_df = df - df.loc[:10].mean(axis=0)

sns.lineplot(data=lam_df, dashes=False)
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```

```python
log_lam_df = np.log(lam_df)

sns.lineplot(data=log_lam_df, dashes=False)
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```
