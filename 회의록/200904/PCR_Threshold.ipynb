# Real-Time PCR

- 실시간 중합효소연쇄반응(영어: real-time polymerase chain reaction, real-time PCR, quantitative real time polymerase chain reaction, qPCR)은 분자생물학에서 중합효소 연쇄 반응을 기반으로한 실험방법이다. 실시간 중합효소연쇄반응은 목표 DNA분자의 증폭과 양의 측정을 동시에 한다. DNA샘플에서 하나 또는 그 이상의 특정 서열을 위해, 실시간 중합효소연쇄반응은 검출과 양의 측정을 할 수 있게한다. 계량은 절대적인 복제 수 또는 상대적인 양을 셀 수 있다.

- DNA 증폭 ->  DNA 검출후 양 측정

- 편의를 위해서 임의로 PCR의 과정을 DNA를 증폭시키는 '증폭부', 증폭시킨 DNA를 검출하는 '검출부'로 나누겠음

### PCR Cycle
위사진 참고

1. DNA는 애초에 이중나선 구조를 띄고 있음
2. 증폭부에서 DNA의 이중나선 구조를 풀기 위해서 94도 정도로 내부온도를 높이게 됨
3. 그 후 55도로 온도를 낮추어 Primer(시발체)가 단일 DNA 끝단에 붙게됨
4. 다시 72도로 온도를 올렸을 때 Taqman Probe(중합효소)가 DNA가닥을 복제하게 되어 완전한 이중나선 구조로 돌아가 DNA가 복제 됨

1~4번과정을 PCR Cycle 이라고 부르고, Cycle과정을 반복하여 한 Cycle당 DNA양이 이론상 2배씩 증폭됨

### PCR 증폭과정에서 형광물질이 발생하는 과정

1. 단일DNA에 형광을 띄는 'Reporter'와 Reporter가 형광을 띄지 못하게 억제하는 'Quencher'가 붙어있음

# 분석

### 분석전에 알아야할 용어들
- Baseline(Background, Noise) : 3~15 정도의 Cycle에서 나타나는 형광신호
- Threshold : Baseline과는 다르게 증폭신호의 증가 양상이 뚜렷하게 구분되는 시점
- Ct(threshold Cycle) : 증폭곡선과 Thresholdline의 교차점(PCR 반응에서 target농도의 상대적 측정값)

Threshold 값을 구하고, Abi회사의 Threshold를 기준으로 비교하는 과정이 필요함

### Cycle threshold 계산함수

* ABI Ct 계산 방법을 분석한 결과 $log(\Delta R_n)$ 을 predetermined threshold로 나누는 것으로 확인 되었다.
 * $R_n = \frac{reporter~dye~fl}{passive~reference~fl}$ (PCR에서 나온 Data)
 * $\Delta R_n = R_n - baseline$
 * Ct는 threshold를 이용하여 아래 그림과 같이 interpolation해서 구한다.
 
 <img src='img/ctCalc.jpg'/>
 
 * 그러므로 같은 방법을 사용하고, 형광 밝기의 상대적인 차이를 알 수 없으니까 parameter를 1개 정해서 그 값을 구한다.
 * $10^{-1}$~$10^{-5}$ 농도에 대한 ABI 장비에서 구한 Ct 값 (2번 실험 평균)(18.2, 21.7, 25.2, 29.0, 32.4)와 가장 가깝도록 parameter를 정한다.

위그림 참고

Linear interpolation : https://ko.wikipedia.org/wiki/%EC%84%A0%ED%98%95_%EB%B3%B4%EA%B0%84%EB%B2%95

$\frac{id-C_t}{id - (id - 1)} = \frac{(d-thres)}{D}$

$C_t = id - \frac{(d-thres)}{D}$

### threshold 결정
- ABI에서 log(fl)값의 threshold는 0.697 이고 이때 fl max는 10 이었다.
- Xavier의 range가 다르므로 상기 그림을 참조해서 3~6까지 바꾸어 가면서 ABI에서 구한 값과 가장 가까운 값을 주는 threshold를 선택한다

````python
def calculateCt(lDb, lth): # lDb는 log(data), lth=log threshold 3~6
    global ct
    ct = np.zeros(len(ct))
    ic=0
    for c in lDb.columns:
        tmp=lDb.loc[lDb[c]>lth]
        if tmp.empty: # lv 가 최대 cycle 보다 크면
            ct[ic]=-1
        else:
            id = tmp[c].index[0]
            rv = lDb[c].loc[id]
            lv = lDb[c].loc[id-1]
            dv = rv-lv
            ct[ic] = id - (rv-lth)/dv
        ic=ic+1
    return ct
````

````python
# Xavier에서의 FAM fluorescence의 최적 threshold 찾기
import matplotlib.pyplot as plt
#ax=plt.gca()
ax = plt.figure(figsize = (10,8))
plt.xlabel('Log threshold')
plt.ylabel('Distance')
#ax.set_xlim(-0.5, 0.5)
plt.ylim(0.705, 0.735)

# Chlamydia Trachomatis (CT data)에서 튀는 값을 제외하고 나머지의 최적 threshold를 구해보자
ct=np.zeros(len(lDb_FAM.columns))
ltharr=np.linspace(4.5,5.5,100) # search area
# 해당 FAM 형광 비교 AbiCt값 만 추출해서 적용
abiCtEx=np.concatenate((abiCt[[0,1,2,3]],abiCt[[0,2,3]],abiCt[[0,1,3,4]],abiCt[[0,1,2,3]]),axis = 0)
dist = np.zeros(len(ltharr))
ic = 0
for th in ltharr:
    ct=calculateCt(lDb_FAM,th)
    dist[ic]=np.std(ct-abiCtEx) # 튀지 않는 data 와 true data의  Ct 차이
    ic = ic+1
midx = np.argmin(dist)
plt.plot(ltharr, dist)
plt.plot(ltharr[midx],dist[midx],'bo')
print(ltharr[midx],min(dist))
plt.savefig("threhold_FAM.png", dpi = 600)
````

````python
# Xavier에서의 HEX fluorescence의 최적 threshold 찾기
import matplotlib.pyplot as plt
ax = plt.figure(figsize = (10,8))
plt.xlabel('Log threshold')
plt.ylabel('Distance')
#ax.set_xlim(-0.5, 0.5)
plt.ylim(1.125, 1.17)

# Neisseria Gonorrhoeae (NG data)에서 튀는 값을 제외하고 나머지의 최적 threshold를 구해보자
ct=np.zeros(len(lDb_HEX.columns))
ltharr=np.linspace(4.5,5.5,100) # search area
# 해당 FAM 형광 비교 AbiCt값 만 추출해서 적용
abiNGEx=np.concatenate((abiNG[[0,1,2,3]],abiNG[[0,1,2,3]],abiNG[[3,4]],abiNG[[0,2,3]]),axis = 0)
dist = np.zeros(len(ltharr))
ic = 0
for th in ltharr:
    ct=calculateCt(lDb_HEX,th)
    dist[ic]=np.std(ct-abiNGEx) # 튀지 않는 data 와 true data의  Ct 차이
    ic = ic+1
midx = np.argmin(dist)
plt.plot(ltharr, dist)
plt.plot(ltharr[midx],dist[midx],'bo')
print(ltharr[midx],min(dist))
plt.savefig("threshold_HEX.png", dpi = 600)
````
