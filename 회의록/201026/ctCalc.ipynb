```python
#Function to calculate Ct
def calcCt(lDb, lth):
    ct = np.zeros(len(lDb.columns))
    ic = 0
    
    lDb_over10 = lDb[10:]
    
    for c in lDb.columns:
        tmp=lDb_over10.loc[lDb_over10[c]>lth]
        
        if tmp.empty:
            ct[ic]=-1
        else:
            id = tmp[c].index[0]
            rv = lDb[c].loc[id]
            lv = lDb[c].loc[id-1]
            dv = rv-lv
            ct[ic] = id - (rv-lth)/dv
        ic += 1
        
    return ct
```

```python
abiCt=np.array([20.868,23.851,27.335,30.499,34.528])
ltharr=np.linspace(4.5, 6,1000)
```

```python
# Get targets matching pcr_df
ct_target = []

for col in pcr_df.columns:
  if col[6] == '5':
    ct_target.append(abiCt[0])
  elif col[6] == '4':
    ct_target.append(abiCt[1])
  elif col[6] == '3':
    ct_target.append(abiCt[2])

ct_target = np.array(ct_target)
```

```python
#Get optimized threshold value for all data
dist = []
for th in ltharr:
    ct = calcCt(pcr_loglam_df, th)
    dist.append(np.mean(ct-ct_target))

dist = np.array(dist)

midx = np.argmin(dist)

print(f'min_error:{np.min(dist)}, thresholdValue:{ltharr[np.argmin(dist)]}, {midx}')

plt.figure(figsize=(10, 10))
plt.xlabel('Log threshold')
plt.ylabel('Distance')

plt.plot(ltharr, dist)
plt.plot(ltharr[midx],dist[midx],'bo')
plt.show()

optim_thres = ltharr[np.argmin(dist)]
```

```python
#Get threshold value 10^5
dist = []
for th in ltharr:
    ct = calcCt(pcr_loglam_df[[ col for col in pcr_df.columns.to_list() if col[6] == '5']], th)
    dist.append(np.std(ct-abiCt[0]))

dist = np.array(dist)

midx = np.argmin(dist)

print(f'min_error:{np.min(dist)}, thresholdValue:{ltharr[np.argmin(dist)]}, {midx}')

plt.figure(figsize=(10, 10))
plt.xlabel('Log threshold')
plt.ylabel('Distance')

plt.plot(ltharr, dist)
plt.plot(ltharr[midx],dist[midx],'bo')
plt.show()
```

```python
#Get threshold value 10^4
dist = []
for th in ltharr:
    ct = calcCt(pcr_loglam_df[[ col for col in pcr_df.columns.to_list() if col[6] == '4']], th)
    dist.append(np.std(ct-abiCt[1]))

dist = np.array(dist)

midx = np.argmin(dist)

print(f'min_error:{np.min(dist)}, thresholdValue:{ltharr[np.argmin(dist)]}, {midx}')

plt.figure(figsize=(10, 10))
plt.xlabel('Log threshold')
plt.ylabel('Distance')

plt.plot(ltharr, dist)
plt.plot(ltharr[midx],dist[midx],'bo')
plt.show()
```

```python
#Get threshold value 10^3
dist = []
for th in ltharr:
    ct = calcCt(pcr_loglam_df[[ col for col in pcr_df.columns.to_list() if col[6] == '3']], th)
    dist.append(np.std(ct-abiCt[2]))

dist = np.array(dist)

midx = np.argmin(dist)

print(f'min_error:{np.min(dist)}, thresholdValue:{ltharr[np.argmin(dist)]}, {midx}')

plt.figure(figsize=(10, 10))
plt.xlabel('Log threshold')
plt.ylabel('Distance')

plt.plot(ltharr, dist)
plt.plot(ltharr[midx],dist[midx],'bo')
plt.show()
```

```python
ct = calcCt(pcr_loglam_df, optim_thres)
ct_error= np.abs(ct-ct_target)
```

```python
pcr_df.loc[40] = ct_target       #target
pcr_df.loc[41] = ct           #output
pcr_df.loc[42] = ct_error   #abs(ct-ct_target)

pcr_df.to_csv('./ct_result.csv', sep=',', na_rep='NaN')
```
